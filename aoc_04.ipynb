```python
TEST = False
DEBUG = False
X = """7,4,9,5,11,17,23,2,0,14,21,24,10,16,13,6,15,25,12,22,18,20,8,19,3,26,1

22 13 17 11  0
 8  2 23  4 24
21  9 14 16  7
 6 10  3 18  5
 1 12 20 15 19

 3 15  0  2 22
 9 18 13 17  5
19  8  7 25 23
20 11 10 24  4
14 21 16 12  6

14 21 17 24  4
10 16 15  9 19
18  8 23 26 20
22 11 13  6  5
 2  0 12  3  7"""

boardWidth = 0

markerBoards = []

if TEST:
    alldata = X.strip().split("\n")
    draw = list(map(int, alldata[0].split(","))) 
    boards  = []
    board = []
    for raw_row in alldata[2:]:
        print( raw_row )
        if raw_row == "":
            boards.append(board)
            markerBoards.append( [0,]*len(board) )
            board = []
        else:
            board.extend( list(map(int, raw_row.split())) )
            if boardWidth == 0:
                boardWidth = len(raw_row.split())
    boards.append(board)
    markerBoards.append( [0,]*len(board) )
    
else:
    with open("aoc_4_input.txt", "r") as f:
        alldata = f.read().strip().split("\n")
        draw = list(map(int, alldata[0].split(","))) 
        boards  = []
        board = []
        for raw_row in alldata[2:]:
            if raw_row == "":
                boards.append(board)
                markerBoards.append( [0,]*len(board) )
                board = []
            else:
                board.extend( list(map(int, raw_row.split())) )
                if boardWidth == 0:
                    boardWidth = len(raw_row.split())
        boards.append(board)
        markerBoards.append( [0,]*len(board) )
```

```python
#Part1
import sys

def printBoards():
    for boardIndex, board in enumerate(boards):
        for cellIndex, mask in enumerate( markerBoards[boardIndex] ):
            if mask == True:
                sys.stdout.write( "|{:>2}| ".format(board[cellIndex]) )
            else:
                sys.stdout.write( " {:>2}  ".format(board[cellIndex]) )
            if cellIndex%5 == 4 :
                print()
        print()

def markBoards( number ):
    for index, board in enumerate(boards):
        if number in board:
            loc = board.index(number)
            markerBoards[index][loc] = 1

def checkBoards( index, markerBoard ):
    for i in range(0, boardWidth):
        #Column check
        columnSum = 0
        for j in range(0,boardWidth):
            columnSum += markerBoard[ i+(j*boardWidth) ]
        if columnSum == boardWidth:
            columnValue = 0
            for j in range(0,boardWidth):
                columnValue += boards[index][ i+(j*boardWidth) ]
            print("Bingo column: ", i, columnSum, columnValue)
            return columnValue
        
        #Rowcheck
        rowSum = sum(markerBoard[(i*boardWidth):(i*boardWidth)+boardWidth])
        if rowSum == boardWidth:
            rowValue = sum(boards[index][(i*boardWidth):(i*boardWidth)+boardWidth])
            print("Bingo row: ", i, rowSum, rowValue)
            return rowValue
    return False

def calcBoard( boardIndex ):
    retValue = 0
    for cellIndex, value in enumerate(boards[boardIndex]):
        if markerBoards[boardIndex][cellIndex] == 0:
            retValue += value
            print("Value", value, retValue)
    print("Value", retValue)
    return retValue

bingo = False
for number in draw:
    markBoards( number )
    for index, markerBoard in enumerate(markerBoards):
        bingo = checkBoards(index, markerBoard)
        if bingo != False:
            print("BINGO: ", index, bingo)
            break
        
    if bingo:
        print(index)
        print(number)
        value = calcBoard( index )
        print()
        print( "Answer: ", number, value, number * value)
        break
```
